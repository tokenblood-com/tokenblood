```python
from datasets import load_dataset
from dotenv import load_dotenv
import os

load_dotenv()
assert "HF_TOKEN" in os.environ, "HF_TOKEN is not set"

ds = load_dataset("Isotonic/pii-masking-200k")
ds
```

```python
ds = ds.filter(lambda item: item["language"] == "en")
ds = ds.remove_columns(["span_labels", "bio_labels", "tokenised_text", "privacy_mask"])
ds = ds.filter(
    lambda item: "FIRSTNAME" in item["masked_text"]
    or "LASTNAME" in item["masked_text"]
    or "MIDDLENAME" in item["masked_text"]
)
ds
```

```python
import random

for ii in random.sample(range(len(ds["train"])), 100):
    print(ds["train"][ii]["unmasked_text"])
    print(ds["train"][ii]["masked_text"])
    print()
```

```python
new_ds = ds.filter(
    lambda item: "FIRSTNAME_2" in item["masked_text"]
    or "LASTNAME_2" in item["masked_text"]
    or "MIDDLENAME_2" in item["masked_text"]
)
new_ds
```

```python
import random

for ii in random.sample(range(len(new_ds["train"])), 10):
    print(new_ds["train"][ii]["unmasked_text"])
    print(new_ds["train"][ii]["masked_text"])
    print()
```

```python
middle_name_ds = ds.filter(lambda item: "MIDDLENAME" in item["masked_text"])

for ii in random.sample(range(len(middle_name_ds["train"])), 10):
    print(middle_name_ds["train"][ii]["unmasked_text"])
    print(middle_name_ds["train"][ii]["masked_text"])
    print()
```
